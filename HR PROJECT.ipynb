```python
Step 3: Exploratory Data Analysis (EDA)
Now that we have an idea of what the dataset looks like, let's explore it with visualizations.

1. Employee Attrition Overview
Let's visualize how many employees have left the company (Attrition analysis)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set the style for seaborn
sns.set(style='whitegrid')
```

```python
# Plot the distribution of Attrition
plt.figure(figsize=(6, 4))
sns.countplot(data=df, x='Attrition', palette='pastel')
plt.title('Employee Attrition Overview')
plt.xlabel('Attrition')
plt.ylabel('Count')
plt.show()
```

```python
2. Attrition by Department
Understanding which departments have higher attrition rates can be insightful
```

```python
# Plot Attrition by Department
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='Department', hue='Attrition', palette='Set2')
plt.title('Attrition by Department')
plt.xlabel('Department')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.legend(title='Attrition')
plt.show()
```

```python
3. Monthly Income Distribution
We can look at how the monthly income is distributed among employees
```

```python
# Monthly Income Distribution
plt.figure(figsize=(8, 5))
sns.histplot(df['MonthlyIncome'], kde=True, color='green')
plt.title('Monthly Income Distribution')
plt.xlabel('Monthly Income')
plt.ylabel('Frequency')
plt.show()
```

```python
4. Job Satisfaction Levels
Let's visualize job satisfaction levels across employees.
```

```python
# Job Satisfaction Levels
plt.figure(figsize=(8, 5))
sns.countplot(data=df, x='JobSatisfaction', palette='coolwarm')
plt.title('Job Satisfaction Levels')
plt.xlabel('Job Satisfaction')
plt.ylabel('Count')
plt.show()
```

```python
Step 4: Insights and Recommendations
Based on the visualizations, we can derive several insights:

Attrition Analysis: High attrition rates in certain departments may indicate underlying issues.
Income Analysis: Understanding income distribution helps identify disparities.
Job Satisfaction: Low job satisfaction could correlate with higher attrition.
Correlation Analysis: Identifying strong correlations between features can guide further analysis
```
